```python
import random

suits = ('Hearts','Diamonds','Spades','Clubs')
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
values = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,
          'Queen':10,'King':10,'Ace':11}
playing = True
```

```python
class Card:
    
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
    
    def __str__(self):
        return f'{self.rank} of {self.suit}'
```

```python
class Deck:
    
    def __init__(self):
        self.deck = []  # start with an empty list
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
                
    def __str__(self):
        all_cards = ''
        for card in self.deck:
            all_cards += '\n'+ card.__str__()
        return all_cards
    
    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self): 
        return self.deck.pop()
```

```python
class Hand:
    
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0
        self.aces = 0    # add an attribute to keep track of aces
    
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank]
        if card.rank == 'Ace':
            self.aces += 1
    
    def adjust_for_ace(self):
        if self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
class Player:
    
    def __init__(self):
        self.name = input('Player enter name  ')
        self.chips = 1000
        
    def make_bet(self):
        while True:
            self.bet = int(input('Enter your bet  '))
            if self.bet <= self.chips:
                break
            else:
                print('Wrong input \nTry again...')
                
    def win_bet(self):
        self.chips += self.bet
    
    def loose_bet(self):
        self.chips -= self.bet
```

```python
def hit(deck, hand):
    hand.add_card(deck.deal())
    hand.adjust_for_ace()
    
def show_some(player, dealer):
    print("\nPlayer's cards are\n")
    for card in player:
        print(card.__str__())
    print("\nDealer's cards are\n")
    print(dealer[-1].__str__())
    
def show_all(player,dealer):
    print("\nPlayer's cards are\n")
    for card in player:
        print(card.__str__())
    print("\nDealer's cards are\n")
    for card in dealer:
        print(card.__str__())
        
def player_busts(chip):
    chip.loose_bet()
    print ('\n Player busted')
    print (f"\n{chip.name} has lost a bet of {chip.bet}\nTotal chips remaining are  {chip.chips}")
    

def player_wins(chip):
    chip.win_bet()
    print ('\nPlayer has won')
    print (f"\n{chip.name} has won a bet of {chip.bet}\nTotal chips remaining are  {chip.chips}")
    
    
def dealer_busts(chip):
    chip.win_bet()
    print('\n Dealer busted')
    print (f"\n{chip.name} has won a bet of {chip.bet}\nTotal chips remaining are  {chip.chips}")
    
    
def dealer_wins(chip):
    chip.loose_bet()
    print ('\nDealer has won')
    print (f"\n{chip.name} has lost a bet of {chip.bet}\nTotal chips remaining are  {chip.chips}")
    
    
def push():
    print("\nIT'S A TIE, NO ONE WINS THIS GAME")
```

```python
game_mode = True
round_num = 1
print ('\nHello welcome to the BLACKJACK game')
player = Player()
while game_mode:
    print('Round ',round_num)
    player.make_bet()
    play_deck = Deck()
    play_deck.shuffle()
    
    player_hand = Hand()
    dealer_hand = Hand()
    
    for turn in range(2):
        player_hand.add_card(play_deck.deal())
        dealer_hand.add_card(play_deck.deal())
    
    show_some(player_hand.cards,dealer_hand.cards)      
    while playing:
        if player_hand.value == 21 and dealer_hand.value != 21:
            player_wins(player)
            break
            
        choice = input("You want to hit or stand?  \nEnter 'h' for hit and 's' for stand   ")
        
        if choice == 'h':            
            hit(play_deck,player_hand)
            show_some(player_hand.cards,dealer_hand.cards)
            
            if player_hand.value > 21:
                player_busts(player)
                playing = False
            else:
                pass
        
        elif choice == 's':
            show_all(player_hand.cards,dealer_hand.cards)
            while dealer_hand.value < 17:
                hit(play_deck,dealer_hand)
                show_all(player_hand.cards,dealer_hand.cards)
            if dealer_hand.value > 21:
                dealer_busts(player)
                break
            elif dealer_hand.value < player_hand.value:
                player_wins(player)
                break
            elif dealer_hand.value > player_hand.value:
                dealer_wins(player)
                break                
            else:    
                push()
                break
                
        else:
            print('\nINVALID INPUT, TRY AGAIN')
            
    while True:
        replay = input('\nWANT TO PLAY AGAIN?\nENTER YES OR NO  ')
        if replay.upper() == 'NO':
            game_mode = False
            break
        elif replay.upper() == 'YES' and player.chips > 0:
            playing = True
            break
        else:
            print("\nYOU HAVE ENTERED WRONG CHOICE\nTRY AGAIN")
    round_num += 1
```
