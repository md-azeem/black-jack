```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return f'{self.rank} of {self.suit}'
```

```python
class Deck:
    
    def __init__(self):
        self.deck = []
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
                
    def shuffle(self):
        random.shuffle(self.deck)
        
    def show(self):
        for card in self.deck:
            print (card)
        
    def deal(self):
        return self.deck.pop()
```

```python
class Player:
    
    def __init__(self, name):
        self.name = name
        self.deck = []
    
    def add_cards(self,cards):
        if type(cards) == list:
            self.deck.extend(cards)
        else:
            self.deck.append(cards)
            
    def remove_cards(self):
        return self.deck.pop(0)
    
    def __str__(self):
        return f'{self.name} has {len(self.deck)}'
```

```python
player1 = Player('Player1')
player2 = Player('Player2')
cards = Deck()
cards.shuffle()
for index in range(26):
    player1.add_cards(cards.deal())
    player2.add_cards(cards.deal())
print ('Game is starting...\n')
round_num = 1
game_mode = True
while game_mode:
    print (f'\nRound {round_num}')
    if player1.deck == []:
        print ('Player1 out of cards \nPlayer2 wins...!')
        break
    if player2.deck == []:
        print ('Player2 out of cards \nPlayer1 wins...!')
        break
    table_card1 = []
    table_card1.append(player1.remove_cards())
    table_card2 = []
    table_card2.append(player2.remove_cards())
    war = True
    while war:
        print ('Player1 has  ',table_card1[-1])
        print ('Player2 has  ',table_card2[-1])
        if table_card1[-1].value > table_card2[-1].value:
            player1.add_cards(table_card1)
            player1.add_cards(table_card2)
            print ('Player1 wins this round')
            war = False
        elif table_card1[-1].value < table_card2[-1].value:
            player2.add_cards(table_card2)
            player1.add_cards(table_card1)
            print ('Player2 wins this round')            
            war = False
        else:
            print ('War will take place')
            if len(player1.deck) <= 13:
                print ('Player1 has insufficient cards for war, Player2 wins!')
                war = False
                game_mode = False
            if len(player2.deck) <= 13:
                print ('Player2 has insufficient cards for war, Player1 wins!')
                war = False
                game_mode = False
            else:
                for i in range(13):
                    table_card1.append(player1.remove_cards())
                    table_card2.append(player2.remove_cards())
                    
    round_num += 1
                
```
